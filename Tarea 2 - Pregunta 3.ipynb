```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import Dense, Conv2D, BatchNormalization, Activation, LeakyReLU, Dropout, Reshape, UpSampling2D, Flatten
from tensorflow.keras.optimizers import RMSprop, Adam
from tensorflow.keras.models import Sequential

import matplotlib.pyplot as plt
import numpy as np
import random

(x, y),(_, _) = mnist.load_data()
x_aux = []
for equis, ye in zip(x, y):
  if ye == 6:
    x_aux.append(equis)

x = np.array(x_aux).reshape((-1,28,28,1)).astype(np.float32)
x = x/255

dropout_prob = 0.4

generator_net = Sequential()

generator_net.add(Dense(7*7*256, input_dim=100))
generator_net.add(BatchNormalization(momentum=0.9))
generator_net.add(LeakyReLU())
generator_net.add(Reshape((7,7,256)))
generator_net.add(Dropout(dropout_prob))

generator_net.add(UpSampling2D())
generator_net.add(Conv2D(128, 5, padding='same'))
generator_net.add(BatchNormalization(momentum=0.9))
generator_net.add(LeakyReLU())

generator_net.add(UpSampling2D())
generator_net.add(Conv2D(64, 5, padding='same'))
generator_net.add(BatchNormalization(momentum=0.9))
generator_net.add(LeakyReLU())

generator_net.add(Conv2D(32, 5, padding='same'))
generator_net.add(BatchNormalization(momentum=0.9))
generator_net.add(LeakyReLU())

generator_net.add(Conv2D(1, 5, padding='same'))
generator_net.add(Activation('sigmoid'))


discriminator_net = Sequential()

discriminator_net.add(Conv2D(64, 5, strides=2, padding='same'))
discriminator_net.add(LeakyReLU())

discriminator_net.add(Conv2D(128, 5, strides=2, padding='same'))
discriminator_net.add(LeakyReLU())
discriminator_net.add(Dropout(dropout_prob))

discriminator_net.add(Conv2D(256, 5, strides=2, padding='same'))
discriminator_net.add(LeakyReLU())
discriminator_net.add(Dropout(dropout_prob))

discriminator_net.add(Conv2D(512, 5, strides=1, padding='same'))
discriminator_net.add(LeakyReLU())
discriminator_net.add(Dropout(dropout_prob))

discriminator_net.add(Flatten())
discriminator_net.add(Dense(1))
discriminator_net.add(Activation('sigmoid'))

optimizer_discriminator = RMSprop(lr=0.0008, clipvalue=1.0, decay=1e-10)
discriminator = Sequential()
discriminator.add(discriminator_net)
discriminator.compile(loss='binary_crossentropy', optimizer = optimizer_discriminator, metrics=['accuracy'])

optimizer_gan = Adam(lr=0.0004, clipvalue=1.0, decay=1e-10)
gan = Sequential()
gan.add(generator_net)
for layer in discriminator_net.layers:
    layer.trainable = False
gan.add(discriminator_net)
gan.compile(loss='binary_crossentropy', optimizer=optimizer_gan, metrics=['accuracy'])

save_results_every = 4
plot_every = 32
batch_size = 128
total_epochs = 256


diss = []
genn = []

from tqdm import tqdm

for i in range(total_epochs):
  images_train = x[np.random.randint(0, x.shape[0], size=batch_size), :, :, :] # select batch_size random images from x
  noise = np.random.uniform(-1.0, 1.0, size=[batch_size, 100]) # generate (batch_size,latent_dim) sized noisenoise
  images_fake = generator_net.predict(noise)

  x_train = np.concatenate((images_train, images_fake))
  y = np.ones([2*batch_size, 1])
  y[batch_size:, :] = 0

  response_dis = discriminator.train_on_batch(x_train, y)

  y_gen = np.ones([batch_size, 1])
  noise_gen = np.random.uniform(-1.0, 1.0, size=[batch_size, 100])
  response_gen = gan.train_on_batch(noise_gen, y_gen)

  if i%save_results_every==0:
    # save accuracies and losses for plotting
    diss.append(response_dis)
    genn.append(response_gen)
    if i%plot_every==0:
      print(i)
      fig, axs = plt.subplots(1, 5, True , True, figsize=(8, 8))
      for n, img in enumerate(images_fake[:50:10]):
        img = images_fake[0]
        axs[n].imshow(img.reshape((img.shape[0], img.shape[1])), cmap='gray') # plot random generated images and losses and accuracies
      plt.show()

plt.show()
plt.plot(diss)
plt.title("Discriminator")
plt.legend(discriminator.metrics_names)
plt.show()
plt.plot(genn)
plt.title("Generator")
plt.legend(gan.metrics_names)
plt.show()
```

**GAN que genere imágenes a color a partir de imágenes en blanco y negro**
Un acercamiento a este tipo de problema es hacer al generador crear imagenes con 3 canales, y hacer que el discriminador discrime esta imagen generada de 3 canales con la base de datos de imágenes RGB. La diferencia estaría en que en este caso el input del generador no sería ruido aleatorio, sino que serían las imágenes a ser coloreadas. Esto obligaría al generador a incluir un esquema encoder-decoder para poder extraer características de la imagen en blanco y negro, para luego incluir un esquema decoder que tome estas características y además añada los canales nuevos. Y de esta forma, el discriminador se encarga de ver que la imagen generada tenga los colores correctos. 

**GAN para fotografías y cuadros**
Para este caso usaría un principio similar al anterior, una imagen como input del generador como esquema encoder-decoder, y otra imagen como dato real para el discriminador. En el caso que se quieran transformar fotos a cuadros, sería necesario entrenar una GAN con fotos como input del generador, y con imagenes de cuadros como datos reales para el driscriminador. Para el caso de que se quieran transformar cuadros a fotos, sería al revés, es decir, entrenar otra GAN con imágenes de cuadros como input del generador, y fotos como datos reales para el discriminador. En estos casos, es el discriminador quien se encarga de "transferir" el estilo de los datos reales a los artificiales (realmente no los transfiere, pero obliga al generador a aplicar ese estilo si quiere engañarlo), mientras que el generador con el esquema encoder-decoder se encarga de construir imágenes que ya no se basan en ruido aleatorio, sino en imágenes concretas.
